# Reading the data

Now the party rocks! When we know how can we obtain the data, it is not a problem to obtain the whole database with all the data from the web-page.
In this lab you will collect your database of Zalando products. You select which goods you want to track. You can define as many filters to your data as you want. Just make sure that the data represents the filters.




```python
import json
import requests
import pandas as pd
from pandas.io.json import json_normalize
```

```python
# Paste the url you obtained for your data

url='https://www.zalando.fr/api/catalog/articles?categories=promo-enfant&limit=84&offset=84&sort=popularity'
```

#### Collect first 84 object of the of the data (1st page)

Your output should be a Pandas DataFrame of goods. Each row should contain only text or numbers, having *family_articles, flags, media* and *sizes* remaining lists (they are exceptions). Hint: use the headers parameter to get the data!

```python
# Headers definition

# Hay que enviar el user-agent para que te devuelva los resultados. Es como hacer un fake del browser.

headers = {'user-agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.87 Safari/537.36'}
```

```python
# Your code

response = requests.get(url, headers = headers)
results = response.json()
results
```

```python
# Después del GET request, tenemos que normalizar los datos obtenidos.

flattened_data = json_normalize(results)

flattened_data.head()
```

```python
# La columna Articles es aún nested. Necesitamos normalize de nuevo.
# Elegimos el primer elemento (0). 

flattened_data_articles = json_normalize(flattened_data['articles'][0])
flattened_data_articles
```

#### Collect all the objects from selected filters. Total number of pages can be found in the same json. Use *sku* column as index.

Your output should be a Pandas DataFrame of goods. Each row should contain only text or numbers, having family_articles, flags, media and sizes remaining lists (they are exceptions).

```python
# Get the total number of pages

total_pages = results["pagination"]["page_count"]
total_pages
```

```python
# Para el total de objects, entiendo que se refiere al total que hay en las 268 páginas.
# No he sabido cómo hacerlo para todas las páginas.
# Hago los siguientes puntos sobre la misma página actual y partiendo del flattened_data_articles que obtuve.

total_articles = results['total_count']
total_articles
```

```python
flattened_data_articles = flattened_data_articles.set_index("sku")
flattened_data_articles
```

#### Display the trending brand in DataFrame

```python
# your code

brands_df = pd.DataFrame(flattened_data_articles["brand_name"].value_counts())
brands_df.head()
```

```python
# Trending Brand será la brand que se repita más veces en esta página. En este caso 'Nike Sportswear'
```

#### Display the brand with maximal total discount (sum of discounts on all goods)

```python
# Our data is still text. Convert prices into numbers.

# your code

# Total discount será la diferencia entre price.original y price.promotional
# Convertimos primero price.original y price.promotional a numbers. 
# Formato que tienen es XX,XX €
# Le quitaremos € y cambiaremos la coma por un punto.

flattened_data_articles['price.original'] = flattened_data_articles['price.original'].str.replace(',','.').str.replace('€','').astype(float)

flattened_data_articles['price.original'].head(5)
```

```python
flattened_data_articles['price.promotional'] = flattened_data_articles['price.promotional'].str.replace(',','.').str.replace('€','').astype(float)

flattened_data_articles['price.promotional'].head(5)
```

```python
# Calculamos cuál sería esta diferencia entre price.original y price.promotional para cada caso.

flattened_data_articles["total_discount"] = flattened_data_articles["price.original"] - flattened_data_articles["price.promotional"]

flattened_data_articles["total_discount"].head(5)
```

```python
# Ahora agruparemos por cada brand_name la suma del total_discount 

total_discount = flattened_data_articles.groupby(['brand_name'])['total_discount'].sum()
total_discount.head(10)
```

```python
# Y ahora lo ordenaremos de mayor a menor para saber la brand con mayor descuento total.

total_discount.sort_values(ascending = False).head()
```

```python
# La marca con mayor descuento total (en esta página) es Nike Sportswear, con 164.30 euros de descuento.
```
